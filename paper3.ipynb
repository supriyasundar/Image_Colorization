```python
from sklearn.metrics import mean_squared_error
from PIL import Image
from numpy import asarray
import numpy as np


import os
path_real = "/content/drive/MyDrive/Georgia Tech/Shared Folders/ECE 6254 Stat ML/test_color/"
path_predicted = "/content/drive/MyDrive/Georgia Tech/Shared Folders/ECE 6254 Stat ML/test_black_colorized/"
%cd /content/
real_list = sorted(os.listdir(path_real))
predicted_list = sorted(os.listdir(path_predicted))
errors = np.zeros(len(real_list))
print(np.shape(errors))
for i in range(len(real_list)):
  print(i)
  real = asarray(Image.open(path_real + real_list[i]))
  predicted = asarray(Image.open(path_predicted + predicted_list[i]))
  error = np.mean((real - predicted) ** 2, dtype=np.float64)
  errors[i] = error
  print(error)
```

```python
print(np.mean(errors))
```

```python
path_real = "/content/drive/MyDrive/Georgia Tech/Shared Folders/ECE 6254 Stat ML/test_color/"
path_predicted = "/content/drive/MyDrive/Georgia Tech/Shared Folders/ECE 6254 Stat ML/test_black_colorized/"
%cd /content/
real_list = sorted(os.listdir(path_real))
predicted_list = sorted(os.listdir(path_predicted))
errorsPercent = np.zeros(len(real_list))
print(np.shape(errors))
for i in range(len(real_list)):
  print(i)
  real = asarray(Image.open(path_real + real_list[i]))
  predicted = asarray(Image.open(path_predicted + predicted_list[i]))
  ncomponents = np.shape(real)[0] * np.shape(real)[1] * np.shape(real)[2]
  # print(ncomponents)
  errorPercent = np.sum(np.abs(real - predicted) / 255.0 * 100) / ncomponents;
  errorsPercent[i] = errorPercent
  print(errorPercent)
 
```

```python
print(np.mean(errorsPercent))
```
